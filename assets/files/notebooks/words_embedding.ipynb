## OneHotEncoding 단어 적용 실습

sklearn의 OneHotEncoder class 사용
- 전체 문장을 입력해야 각 단어의 벡터 생성 가능 
- 또한 특정 벡터가 어떤 단어를 표현하는지 get_features_names_out() 메소드로 확인 가능 

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
sentence = '사과는 맛있다 바나나는 맛있다'
```

```python
# 문장을 단어로 분리
# 띄어쓰기 단위를 사용
words = sentence.split()
# split() method default는 띄어쓰기
print(words)
```

```python
# 단어 배열을 Numpy 배열로 변환
words_array = np.array(words).reshape(-1,1)
```

```python
print(words_array)
# 설명:
# 1. 파이썬 리스트와 Numpy 배열의 차이:
#    - 파이썬 리스트: 다양한 타입의 데이터를 저장 가능, 동적 크기, 수치 연산 비효율적.
#    - Numpy 배열: 동일 타입의 데이터 저장, 고정 크기(재할당 가능), 수치 연산에 매우 효율적(C, Fortran 기반 최적화), 머신러닝 라이브러리(sklearn 등)와 호환.
#    - 변환 이유: 대규모 수치 연산의 성능, 메모리 효율성, 라이브러리 호환성.
# 2. `reshape(-1, 1)`과 2차원 배열:
#    - `words_array`의 `[['사과는'] ['맛있다'] ['바나나는'] ['맛있다']]` 형태는 2차원 배열을 의미합니다.
#    - 바깥쪽 `[]`는 배열 전체를, 안쪽 `[]`는 각 행을 나타냅니다.
#    - `reshape(-1, 1)`은 배열을 2차원으로 재구성하며, 열의 개수를 1로 지정합니다.
#    - `-1`은 행의 개수를 자동으로 계산하도록 합니다. (원래 4개의 요소이므로 4행 1열이 됨).
#    - 이 형태로 변환하는 것은 주로 One-Hot Encoding 등 머신러닝 전처리 시 각 특징을 독립적인 열로 처리하기 위함입니다.
```

```python
encoder = OneHotEncoder(sparse_output=False)
# sparse_output은 OneHotEncoder의 결과 형식을 일반 NumPy 배열(False)로 할지, 메모리 효율적인 희소 행렬(True)로 할지 결정하는 속성

one_hot_encoded = encoder.fit_transform(words_array)
print(one_hot_encoded)
```

```python
encoder.get_feature_names_out()
```

```python
def one_hot_encoded_words(sentence):
  words = sentence.split()
  words_array = np.array(words).reshape(-1,1)
  encoder = OneHotEncoder(sparse_output=False)
  one_hot_encoded = encoder.fit_transform(words_array)
  return one_hot_encoded, encoder.get_feature_names_out()
```

```python
example_sentence = '오늘은 날씨가 참 맑습니다.'
```

```python
encoded_words, feature_names = one_hot_encoded_words(example_sentence)
```

```python
encoded_words, feature_names
```

## Word2Vec을 단어에 적용하기

```python
from gensim.models import KeyedVectors
from gensim.downloader import load

def use_word2vec(word, model):
  try:
    word_vector = model[word]
    return word_vector
  except KeyError:
    return '단어가 모델의 어휘에 없습니다.'
```

```python
word2vec_model = load('word2vec-google-news-300')
```

```python
print(use_word2vec('apple', word2vec_model))
print(use_word2vec('instagram', word2vec_model))
# 학습 당시에 존재하지 않던 단어는 임베딩 불가 
```

```python
from scipy.spatial.distance import cosine 

def word_similarity(word1, word2, model):
    try:
        vector1 = model[word1]
        vector2 = model[word2]
        similarity = 1 - cosine(vector1, vector2) #코사인 유사도 계산 
        return similarity
    except KeyError as e: 
        return str(e)
    
def most_similar(word, model, topn=5):
    try:
        similar_words = model.most_similar(word, topn=topn) 
        return similar_words
    except KeyError as e : 
        return str(e)
```

```python
# 두 단어 사이의 유사도 확인 
print('football & basketball 유사도 : ', word_similarity('football','basketball',word2vec_model))
print('football & airplane 유사도 : ', word_similarity('football','airplane',word2vec_model))
```

```python
# 특정 단어와 가장 유사한 단어 보여주기 
print('football과 가장 유사한 단어 5개는 : ' , most_similar('football', word2vec_model, topn=5))
```

### GloVe를 단어에 적용하기

```python
def use_glove(word, model):
    try:
        word_vector = model[word]
        return word_vector
    except KeyError:
        return '단어가 모델의 어휘에 없습니다.'
```

```python
# 제공하는 Glove 모델 불러오기 
# 약 5분 정도 소요 
glove_model = load('glove-wiki-gigaword-300')
```

```python
print(use_glove('apple', glove_model))
print(use_glove('airpods', glove_model)) # 역시 학습 당시에 없던 단어는 임베딩 불가 
```

```python
print('football & basketball 유사도 : ', word_similarity('football','basketball',glove_model))
print('football & airplane 유사도 : ', word_similarity('football', 'airplane', glove_model))
# 이전의 word2vec_model보다 훨씬 더 극명하게 나뉘는 모습을 볼 수 있다. 
# 전체 데이터 안의 문맥적인 정보를 파악하는 모델이기에 다음과 같은 결과가 나온다.
```

```python
# 특정 단어와 가장 유사한 단어 보여주기 (GloVe)
print('football과 가장 유사한 단어 5개는 : ', most_similar('football', glove_model, topn=5))
```
