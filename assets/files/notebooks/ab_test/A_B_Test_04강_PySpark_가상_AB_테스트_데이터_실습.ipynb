## Look at analytics.variant_daily_sessions

```python
user_event.createOrReplaceTempView("user_event")
user_variant.createOrReplaceTempView("user_variant")
user_metadata.createOrReplaceTempView("user_metadata")
```

```python
variant_daily_sessions = spark.sql("""
SELECT
    variant_id,
    ue.user_id,
    datestamp,
    count(distinct item_id) num_of_items, -- 총 impression
    sum(clicked) num_of_clicks,                -- 총 click
    sum(purchased) num_of_purchases,  -- 총 purchase
    sum(paidamount) revenue                   -- 총 revenue
FROM user_event ue
JOIN user_variant uv ON ue.user_id = uv.user_id
GROUP by 1, 2, 3
""")
```

```python
variant_daily_sessions.show()
```

```python
variant_daily_sessions.count()
```

```python
variant_daily_sessions.createOrReplaceTempView("variant_daily_sessions")
```

```python
spark.sql("""SELECT
    variant_id,
    COUNT(1) count,
    AVG(revenue) revenue_mean,
    SUM(revenue) revenue_sum
FROM variant_daily_sessions
GROUP BY 1""").show()
```

### 직접 t-score 계산해보기

 - 그룹별로 크기와 평균과 각 원소 제곱의 합을 알면 계산 가능
   - mean_b와 mean_a (평균)
   - n_b와 n_a (크기)
   - square_b와 square_a (제곱의 평균)
 - 먼저 variance(표준편차의 제곱)은 아래와 같이 계산
   - var_b = square_b - mean_b*mean_b
   - var_a = square_a - mean_a*mean_a
 - 최종 t-score는 아래와 같이 계산
   - (mean_b - mean_a)/math.sqrt(var_b/n_b + var_a/n_a)

```python
spark.sql("""SELECT
    variant_id,
    COUNT(1) count,
    AVG(revenue) revenue_mean,
    VARIANCE(revenue) revenue_variance
FROM variant_daily_sessions
GROUP BY 1""").show()
```

```python
df = spark.sql("""SELECT
    variant_id,
    COUNT(1) count,
    AVG(revenue) revenue_mean,
    VARIANCE(revenue) revenue_variance
FROM variant_daily_sessions
GROUP BY 1""").toPandas()
```

```python
import math

def compute_2samp_ttest(df):
    # t_score = (mean_b - mean_a)/math.sqrt(var_a/n_a+var_b/n_b)
    mean_a = df[df["variant_id"] == 'control']["revenue_mean"].iloc[0]
    mean_b = df[df["variant_id"] == 'test']["revenue_mean"].iloc[0]
    var_a = df[df["variant_id"] == 'control']["revenue_variance"].iloc[0]
    var_b = df[df["variant_id"] == 'test']["revenue_variance"].iloc[0]
    n_a = df[df["variant_id"] == 'control']["count"].iloc[0]
    n_b = df[df["variant_id"] == 'test']["count"].iloc[0]

    return   (mean_b - mean_a)/math.sqrt(var_a/n_a+var_b/n_b)
```

```python
print(compute_2samp_ttest(df))
```
